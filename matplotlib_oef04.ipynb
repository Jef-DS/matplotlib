Hoe kun je de limiet voor de y-as op 101 zetten?
1. ax.ylim(top:101)
1. ax.ylim(right=101)

1. ax.set_ylim(right=101)


````python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.DataFrame({'weekend':[5, 22]}, index=['zaterdag', 'zondag'])
df_percent = df/df['weekend'].sum() * 100
fig = plt.figure()
ax = fig.add_subplot()
ax.bar(df_percent.index, df_percent['weekend'])
ax.text(0, df_percent.loc['zaterdag', 'weekend']-4, f"{df_percent.iat[0, 0]:.0f}", ha="center", color='white')
ax.text(1, df_percent.loc['zondag', 'weekend']-4, f"{df_percent.iat[1, 0]:.0f}", ha="center", color='white')
ax.set_title('Percentage verkoop in het weekend')
#Wat moet je hier invullen?
plt.show()
````
