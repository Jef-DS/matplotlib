````
import matplotlib.pyplot as plt
gemiddelde = iris.groupby('species').mean(numeric_only=True)
std = iris.groupby('species').std(numeric_only=True)
soorten = std.index.values
plt.errorbar(soorten, gemiddelde.sepal_length, std.sepal_length*3,marker='o', linestyle='',
             ecolor='lightgray', elinewidth=3, capsize=4, label='sepal_length')
plt.errorbar(soorten, gemiddelde.sepal_width, std.sepal_width*3,marker='o', linestyle='',
             ecolor='lightgray', elinewidth=3, capsize=4, label='sepal_width')
plt.legend()

plt.show()


import matplotlib.pyplot as plt
import numpy as np
corr_coef = np.corrcoef(iris.loc[iris.species=='setosa', ['sepal_width','sepal_length', 'petal_width', 'petal_length']].T)
fig, ax = plt.subplots(layout='constrained')
im = ax.imshow(corr_coef, cmap='Greys')
cbar = ax.figure.colorbar(im, ax=ax)
ax.set_xticks(range(len(iris.columns[:-1])), labels=iris.columns[:-1])
ax.set_yticks(range(len(iris.columns[:-1])), labels=iris.columns[:-1])
ax.spines[:].set_visible(False)
ax.set_title('Correlatie sepal en petal data voor setosa')

plt.show()
