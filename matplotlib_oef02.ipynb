````python
import matplotlib.pyplot as plt
from matplotlib.dates import DayLocator, DateFormatter
import numpy as np
import pandas as pd
index = pd.date_range('2026-01-01', periods=10)
rng = np.random.default_rng(42)
data = rng.integers(5, 10, size=len(index))
df = pd.DataFrame({'dagen': index, 'aanwezigen': data})
fig = plt.figure(figsize=(5, 3))
ax = fig.add_subplot()
ax.plot(df['dagen'], df['aanwezigen'])
ax.xaxis.set_major_formatter(DateFormatter('%d-%m-%Y'))
# Wat moet je hier invullen?
plt.show()
````

````python
import matplotlib.pyplot as plt
from matplotlib.dates import DayLocator, DateFormatter
import numpy as np
import pandas as pd
index = pd.date_range('2026-01-01', periods=10)
rng = np.random.default_rng(42)
data = rng.integers(5, 10, size=len(index))
df = pd.DataFrame({'dagen': index, 'graden': data})
plt.figure(figsize=(5, 3))
plt.plot(df['dagen'], df['graden'])
plt.gca().xaxis.set_major_formatter(DateFormatter('%d-%m-%Y'))
plt.gca().xaxis.set_major_locator(DayLocator(bymonthday=[1, 5, 10]))
# Wat moet je hier invullen?
plt.show()
````

# Oplossingen
## Oefening 1
plt.gca().spines['bottom'].set_position('zero') (antwoord 3). Omdat we met plt werken, moeten we eerst de achterliggende axes (grafiek) opvragen. Vervolgens kunnen we de bottom-spine (x-as) op het nulniveau zetten.
## Oefening 2
plt.gca().xaxis.set_major_formatter(DateFormatter("%d")) (antwoord 2). Een 'spine' verwijst naar de lijn. Maar hier willen we de waarden bij de tickmarks aanpassen. Dat gebeurt via de 'xaxis' van de huidige grafiek.
## Oefening 3
ax.xaxis.set_major_locator(DayLocator(bymonthday=[1, 5, 10])) (antwoord 1). In dit geval hebben we een verwijzing naar de huidige grafiek dankzij 'ax'. We kunnen de 'xaxis' dus rechtstreeks aanspreken. Welke tickmarks getoond worden, stellen we in met een 'locator'.
## Oefening 4
plt.gca().yaxis.set_major_formatter('{x:.0f}C') (antwoord 2). Je moet 'x' gebruiken als verwijzing naar de waarde bij een StrMethodFormatter.
## Oefening 5
plt.xlim(left=index[0]) (antwoord 2). Een limiet hangt niet samen met tickmarks of teksten bij de tickmarks. Vandaar dat we die rechtstreeks kunnen aanspreken, zonder dat we de current axes moeten opvragen.
